```python
# Practice 3
# Small Exercise: Age Checker

'''Age Checker'''

# Goal: Ask the user for their age

# Use if-else statements to tell them if:
# - they can drive on their own ( bigger than 18 )
# - they need an adult with them (bigger than 16, but smaller than 18)
# - they are not allowed to drive yet ( smaller than 18 )

# 1. Use input() to ask for the user's age
# 2. Use an if-else block to print different messages based on age
# 3. Add a final print statement that runs no matter what


# 1. Ask for the user's age
age = int(input("Enter your age: "))

# 2. Use if-else to check driving eligibility
if age > 18:
    print("You can drive on your own.")
elif age > 16 and age < 18:
    print("You need an adult with you.")
else:
    print("You are not allowed to drive yet.")

# 3. Final statement that always runs
print("Thank you for using the Age Checker!")
```

```python
# Practice 4
# Small Exercise: Number Pattern

'''Number Pattern'''

# Goal: Use a for loop to print numbers from 1 to 10
# Add an extra print statement to show when a number is even or odd

# 1. Use range() in a for loop to loop from 1 to 10
# 2. Print each number
# 3. Use an if - else statement to print "Even number!" if it's even or "Odd number!" if it's odd

# remember we can check whether the number is even if the number % 2 is equal to zero!

# 1. Loop from 1 to 10
for num in range(1, 11):
    # 2. Print the number
    print(num)

    # 3. Check even or odd
    if num % 2 == 0:
        print("Even number!")
    else:
        print("Odd number!")
```

```python
# Practice #5
# Small Exercise: Password Guessing Game

'''Password Guessing Game'''

# Goal: Ask the user to guess a secret password
# Repeat until they guess it correctly
# Count how many tries it took and give feedback

# 1. Set a secret password in a variable
# 2. Use a while loop to keep asking until user input matches it
# 3. Use a counter to track number of attempts
# 4. Use if-else to give different feedback based on correctness

# Inital Values
secret_password = "ai_ml_digilians"
attempts = 0

# 1st guess already asked
user_guess = input("Guess the password: ")
attempts += 1

# 2. Keep asking until correct
while user_guess != secret_password:
    print("Wrong password! Try again.")
    user_guess = input("Guess the password: ")
    attempts += 1

# When the loop ends, the password is correct
print(f"Correct! 🎉 You guessed the password in {attempts} attempts.")
```

```python
# Practice #6
# Small Exercise: Simple Calculator

# Task: Complete the functions below to perform basic math operations
# Handle division by zero using an if statement

def add(x, y):
    print(f"Adding {x} and {y}: ")
    return x + y   # ✅ add

def subtract(x, y):
    print(f"Subtracting {x} and {y}: ")
    return x - y   # ✅ subtract

def multiply(x, y):
    print(f"Multiplying {x} and {y}: ")
    return x * y   # ✅ multiply

def divide(x, y):
    print(f"Dividing {x} and {y}: ")
    if y == 0:
        return "Cannot divide by zero!"   # ❗ handle division by zero

    return x / y    # ✅ divide normally

# Get user input
num1 = float(input("Enter the first number: "))
num2 = float(input("Enter the second number: "))
operation = input("Choose operation (+, -, *, /): ")

# Perform the operation based on user input
if operation == "+":
    print("Result:", add(num1, num2))
elif operation == "-":
    print("Result:", subtract(num1, num2))
elif operation == "*":
    print("Result:", multiply(num1, num2))
elif operation == "/":
    print("Result:", divide(num1, num2))
else:
    print("Invalid operation. Please choose +, -, * or /.")
```

```python
# Practice #10
# Small Exercise: Rate and Update Your Activities

# Task: Complete the code to practice working with lists and loops
# You will:
# 1. Choose your favorite activity
# 2. Update its rating with a new one
# 3. Print all activities and their updated ratings

activities = ["Gaming", "Reading", "Walking", "Watching Movies"]
ratings = [8, 6, 7, 9]

favorite = input("Which activity is your favorite from the list? ")
new_favorite_rating = int(input(f"What rating would you like to give {favorite} now? "))

# TODO: Loop through the activities list and update the rating of the favorite activity

# Update the rating
for i in range(len(activities)):
    if activities[i] == favorite:
        ratings[i] = new_favorite_rating   # update the matching rating

# Print all activities with updated ratings
for i in range(len(activities)):
    print(f"{activities[i]}: {ratings[i]}")
```

```python
# Practice #11
# Small Exercise: Favorite Mobile Apps

# Task:
# 1. Create a dictionary called 'apps' where the keys are names of 2 apps you use often
#    and the values are how many hours you spend on each per week.
#    Example: {, "YouTube": 8, "Snapchat": 2}
# 2. Add one new app and the time you spend on it weekly.
# 3. Update the time spent on your most-used app.
# 4. Loop through the dictionary and print each app and how many hours you use it in the format: "App: X hours/week"

# TODO: Create the dictionary
# 1. Create the dictionary
apps = {
    "Instagram": 5,
    "YouTube": 8
}

# 2. Add a new app and usage time
apps["TikTok"] = 4

# 3. Update the usage time of your most-used app
# find the app with the highest hours
most_used = max(apps, key=apps.get)
apps[most_used] += 2   # example: add 2 more hours

# 4. Loop through and print results
for app, hours in apps.items():
    print(f"{app}: {hours} hours/week")
```

```python
########### Group Exercise #############

# Group Exercise: Student Report Card
# Task: Complete the functions below to build a simple report card system

# Step 1: Create a student record
def create_student_record(name, age, subjects, scores):
    # Use subjects and scores to build a dictionary of grades
    # Example: {"Math": 96, "English": 87, "French": 90}
    # subjects = ['Math', 'English', 'French]
    # scores = [96, 87, 90]
    grades = {}

    # TODO: build this dictionary with a loop
    for i in range(len(subjects)):
            grades[subjects[i]] = scores[i]
    return {
        "name": name,
        "age": age,
        "grades": grades
    }

# Step 2: Calculate the average grade
def calculate_average_grade(student):
    # TODO: find the average of all subject grades
    # Save it as student["avg_grade"]
    grades = student["grades"].values()
    avg = sum(grades) / len(grades)
    student["avg_grade"] = avg
    return student


# Step 3: Assign a letter grade
def assign_letter_grade(student):
    avg = student["avg_grade"]

    # 90–100 → A
    # 80–89  → B
    # 70–79  → C
    # 60–69  → D
    # below 60 → F

    # TODO: write if-elif-else to assign:
    # student["letter_grade"] = ...
    if avg >= 90:
        student["letter_grade"] = "A"
    elif avg >= 80:
        student["letter_grade"] = "B"
    elif avg >= 70:
        student["letter_grade"] = "C"
    elif avg >= 60:
        student["letter_grade"] = "D"
    else:
        student["letter_grade"] = "F"

    return student

# Step 4: Print the student report
def print_student_report(student):
    # TODO: print name, age, subject grades, average, and letter grade
    print(f"Student Name: {student['name']}")
    print(f"Age: {student['age']}")
    print("Grades:")

    for subject, grade in student["grades"].items():
        print(f"  {subject}: {grade}")

    print(f"Average Grade: {student['avg_grade']:.2f}")
    print(f"Letter Grade: {student['letter_grade']}")
    #pass

# Step 5: Create some student records
all_students = []
# Example:
# all_students.append(
#     create_student_record("Laila", 16,
#         ["Math", "Science", "English"],
#         [90, 85, 92])
# )
all_students.append(
    create_student_record(
        "Laila", 16,
        ["Math", "Science", "English"],
        [90, 85, 92]
    )
)
all_students.append(
    create_student_record(
        "Omar", 17,
        ["Math", "History", "English"],
        [75, 80, 70]
    )
)

all_students.append(
    create_student_record(
        "Mariam", 15,
        ["Biology", "Chemistry", "Physics"],
        [95, 89, 93]
    )
)

# Step 6: Process and print each student's report
for student in all_students:
    student = calculate_average_grade(student)
    student = assign_letter_grade(student)
    print_student_report(student)
    print()
```
