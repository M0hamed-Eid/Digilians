## ***What is Pandas Library?***

Pandas is an open-source Python library that provides data structures and data analysis tools for handling and manipulating structured data. It's widely used in data science, data analysis, and machine learning workflows due to its ease of use and powerful capabilities. Pandas excels at handling structured data, such as tabular data (similar to data in spreadsheets or SQL tables), and it offers functionalities similar to those found in SQL or Excel.

**Key features of Pandas Library include:**


1.   **Series:** 1D array and can hold different data types "single row with multiple columns"
2.   **DataFrame:** tabular data structure with labeled and axes "2D array with mutilple rows and columns"
3.   **Manipulation of data**: perform tasks like filtering rows, merging data, handling missing values, and more.
4. **Data loading**: read data from various file formats such as CSV, Excel, SQL databases, and more. This makes it easy to import data from different sources.
5. **Data analysis**: perform exploratory data analysis (EDA), compute descriptive statistics, calculate correlations, and create pivot tables.
6. **Data visualization**:Pandas itself doesn't offer visualization but it can work seamlessly with visualization libraries like Matplotlib and Seaborn to create graphs and charts.




1.   **Series**

```python
import pandas as pd

# Create a Series from a list
data = [10, 20, 30, 40, 50]
s = pd.Series(data)

# Display the Series
print("Series:")
print(s)

# Access values using indexing, ex. access the third element
print("\nAccessing Values:" , s[2])

# Basic statistics
print("\nBasic Statistics:")
print("Mean:", s.mean())
print("Max:", s.max())
print("Min:", s.min())
```


2.   **DataFrame**



```python
import pandas as pd

# create a dataframe
data={
    'Name':['Ahmed','Noha','Ali','Maryam'],
    'Age':[43,19,15,23],
    'Salary':[1500,1200,3000,5000]
}

# Acess the dataframe
df=pd.DataFrame(data)

# Display the dataframe using method display() or just the name of dataframe
print("The DataFrame is:")
display(df)
```

```python
# Access any COLUMN from the dataframe
print('Access a specific column')
print(df['Name'])
print(df['Age'])

# Acess multiple columns
print('\nAcess multiple columns')
df[['Name','Age']]
```

```python
# Add a list of names to give each row a name:
df = pd.DataFrame(data, index = ["user1", "user2", "user3","user4"])
df
```

```python
# Access any ROW from the dataframe using index, when you access element exceed the index, rais an IndexError (out of bounds)
# Two methods can be used for slicing, .loc selects rows and columns by Names/Labels, while .iloc selects rows and columns by Integer Index/Position.

# Select Single Row by Index Label
print(df.loc['user2'])
# Select Single Row by Index
print(df.iloc[1])

print('Access the first row:')
print(df.loc['user1'])
print('--------------------')
print('Access the fourth row:')
print(df.loc['user3'])
print('--------------------')
print('Access range of rows:')
print(df.loc['user2': 'user4'])
print('--------------------')
print('Access a specific row based on attribute:')
df.loc[df.Salary==1500] # if more then one row with the same value, display all
```

```python
# Access specific value from the dataframe
print(df)
# For ex. access the age of Ali (the 2nd column in third row)
print('\nThe Age of Ali',df.iloc[2,1])
```

```python
# calculate some statistics
print("\nStatistics:")
print("The mean of Age is:", df['Age'].mean())
print("The maximum Age is:", df['Age'].max())
print("The minimum Age is:", df['Age'].min())
print("The maximum Salary is",df['Salary'].max())
```

Exploratory Data Analysis (EDA) involves understanding the data, identifying patterns, and gaining insights before conducting more in-depth analyses.

```python
# Calculate correlation matrix that represent the strength between attributes/features
correlation_matrix = data.corr()

print("Correlation Matrix:")
correlation_matrix
```

6. **Data visualization**

**Matplotlib**: is a plotting library for data visualization, it supports various drawing types s. as plots, histograms, bar charts, and other.

**Seaborn**: is an open-source Python data visualization library built on top of Matplotlib. It is specifically designed for creating attractive and informative statistical graphics.

**Based on the type of fatures, various plots are used:**
* For continuous features: Box Plot, Histogram, Density Representation
* For discrete features: Bar Chart, Dot Chart
* For categorical features: Pie Chart is most common. There exist a lot of newer representations, but in the end they all provide the exact same information.

```python
# Matplotlib

##################### Line Plot ###################
import matplotlib.pyplot as plt

# Sample data
x = [1, 2, 3, 4, 5]
y = [10, 15, 7, 12, 8]

# Create a line plot
plt.plot(x, y, marker='*')

# more properites
#plt.plot(x,y,marker='o',markersize=7,linestyle='-.',linewidth='4',color='blue')
#plt.grid(true)

# Add labels and title
plt.xlabel('X-axis')
plt.ylabel('Y-axis')
plt.title('Simple Line Plot X versue Y')

# Display the plot
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data_test = {
    'Year': [2010, 2011, 2012, 2013, 2014],
    'Sales': [500, 800, 1700, 1500, 2000]
}
df = pd.DataFrame(data_test)

# Create a line plot with marker
plt.plot(df['Year'], df['Sales'], marker='o')
plt.xlabel('Year')
plt.ylabel('Sales')
plt.title('Sales Over Years')

# we can observed that year apperas as a floating value, change it to integers
#plt.xticks(df['Year']) #

plt.grid(True)
plt.show()
```

```python
##################### Bar Plot ###################
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame with type category
data_test = {
    'Category': ['A', 'B', 'C', 'D'],
    'Sales': [500, 800, 1200, 1500]
}
df = pd.DataFrame(data_test)

# Create a bar plot
#plt.bar(df['Category'], df['Sales'])

# add many options as color
plt.bar(df['Category'], df['Sales'],color='orange', width=0.5)

plt.xlabel('Category')
plt.ylabel('Sales')
plt.title('Sales by Category')
plt.show()
```

```python
##################### Scatter Plot ###################
import pandas as pd
import matplotlib.pyplot as plt

# Create a DataFrame
data_test = {
    'Age': [25, 30, 22, 35, 28],
    'Income': [5000, 60000, 4500, 70000, 55000]
}
df = pd.DataFrame(data_test)

# Create a scatter plot
plt.scatter(df['Age'], df['Income'])
plt.xlabel('Age')
plt.ylabel('Income')
plt.title('Income vs. Age')
plt.grid(True)
plt.show()
```

```python
##################### Histogram ###################
import pandas as pd
import matplotlib.pyplot as plt

# the histogram represents the number of occurrence of each element
# Create a DataFrame
data_test = {
    'Scores': [85, 78, 92, 65, 80, 75, 88, 90, 82, 70]
}
df = pd.DataFrame(data_test)

# Create a histogram
plt.hist(df['Scores'], bins=10, edgecolor='black', width=1.5, color='red')
plt.xlabel('Scores')
plt.ylabel('Frequency')
plt.title('Distribution of Scores')
plt.show()
```

```python
##################### Box Plot ###################
# The boxplot shows the variance of data, mean and how far.
numbers={10,15,20,35,40,90,35,60,75}
plt.boxplot(numbers)
plt.show()
```
