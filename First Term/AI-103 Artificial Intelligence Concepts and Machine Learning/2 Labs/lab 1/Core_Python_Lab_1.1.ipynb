## What Are Variables in Python?

<br>


### Variable Naming Rules in Python

To keep your code clean and bug-free, follow these rules when naming variables:

<br>

- Must **start with a letter or underscore**  
  ➤ ✅ `valid_name`  
  ➤ ❌ `1invalid` *(can't start with a number)*
<br>

- Names can include **letters**, **numbers**, and **underscores**

<br>

- **Case-sensitive**

#### *You Have Already Used Some Functions!*
The `print()` function for instance, while quick and easy to use in Python, actually has many lines of code hidden inside of it. In fact, most built-in Python functions like print are not even written in Python, they are implemented in another programming language, **C**, which Python itself is built on.

Here is a simplified version of the C source code that powers the `print()` function in Python:
```c
static PyObject *
builtin_print(PyObject *self, PyObject *args, PyObject *kwds)
{
    PyObject *sep = NULL, *end = NULL, *file = NULL, *objects;
    PyObject *result = NULL;
    int i, err;

    static char *kwlist[] = {"sep", "end", "file", "flush", 0};

    /* Parse arguments: objects to print + keyword args */
    if (!PyArg_ParseTupleAndKeywords(args, kwds, "|OOOO:print",
                                     kwlist,
                                     &sep, &end, &file))
        return NULL;

    /* Default values */
    if (sep == NULL)
        sep = PyUnicode_FromString(" ");
    if (end == NULL)
        end = PyUnicode_FromString("\n");
    if (file == NULL)
        file = _PySys_GetObjectId(&PyId_stdout);  /* sys.stdout */

    /* Print each object */
    for (i = 0; i < PyTuple_Size(objects); i++) {
        PyObject *text = PyObject_Str(PyTuple_GetItem(objects, i));
        err = PyFile_WriteObject(text, file, Py_PRINT_RAW);
        Py_DECREF(text);
        if (err)
            return NULL;
        if (i < PyTuple_Size(objects) - 1)
            PyFile_WriteObject(sep, file, Py_PRINT_RAW);
    }

    /* Add the "end" string */
    PyFile_WriteObject(end, file, Py_PRINT_RAW);

    Py_RETURN_NONE;
}
```

When you **call** `print("Hello")`, Python is really **calling** this underlying C function.

It takes your objects, converts them into text, adds spaces or newlines as needed, and writes them to the screen. This shows that built-in functions are not magic, they are just pre-written code in another language.

<br>

### Indexing and Slicing in Python Lists

#### Indexing

- Each item in a list has a position number called its **index**
- Indexes start from `0` for the first item
- You can access individual items using square brackets:

```python
names = ["Ali", "Laila", "Sara"]
print(names[0])  # Ali
print(names[1])  # Laila
print(names[2])  # Sara
```

- You can also use negative indexes to count from the end:

```python
print(names[-1])  # Sara (last item)
print(names[-2])  # Laila (second to last)
```
<br>

#### Slicing

- Slicing lets you access a range of items using the format:
```python
list[start:end] #gives you items starting from start up to but not including end
```
- If the **first argument** in the slice is left empty, it means "start from the beginning".  
- If the **second argument** is left empty, it means "go until the end".
```python
names = ["Ali", "Laila", "Sara", "Omar", "Hana"]
print(names[1:4])   # ['Laila', 'Sara', 'Omar']
print(names[:3])    # ['Ali', 'Laila', 'Sara'] (from start up to index 3)
print(names[2:])    # ['Sara', 'Omar', 'Hana'] (from index 2 to the end)
```

#### Index Out of Range

If you try to access an index that does not exist in the list, Python will give an error called **`IndexError`**.

```python
names = ["Ali", "Laila", "Sara"]
print(names[3])    # ❌ IndexError: list index out of range
print(names[-3])   # ❌ IndexError: list index out of range
```

## **1.8 Working with Dictionaries**

Dictionaries are **unordered** collections of **key-value pairs**. They allow you to associate unique keys with specific values, making them useful for structured data.

Each **key** must be unique (like a string or number), while the **value** can be any data type (even lists).


```python
dictionary = {
    "key1": "value1",
    "key2": "value2,
    "key3": "value3,
}

# dictionaries are simillar to lists with iterative int keys
list_like_dict = {
    0: "first item",
    1: "second item",
    2: "third item",
}
```

| Feature              | List                                 | Dictionary                              |
|-----------------------|----------------------------------------|--------------------------------------------|
| How it stores things  | In order, like a row of numbered boxes | With labels (**keys**) you choose for each box |
| How to get an item    | Use its position number (**index**)        | Use its label (**key**)                        |
| Example               | `myList = ["a","b","c"]`               | `myDict = {"x":"a","y":"b","z":"c"}`       |
| Getting an item       | `myList[0]` → `"a"`                    | `myDict["x"]` → `"a"`                      |
| Keys/Indexes          | Always numbers starting at 0           | Can be numbers, words, or other things     |
| When to use           | If order matters and numbers are fine  | If you want custom labels for your items   |

#### The key difference

- **Lists** `[]`: you always use numbers that go in order (0, 1, 2, …).

- **Dictionaries** `{}`: you can choose any label you want, not just numbers. For example:


````python
old_way = ["first item", "second item", "third item"]
print(old_way[0])

new_way = {0: "first item", 1: "second item", 2: "third item"}
print(new_way[1])
````

## Intro to Classes
So far, you've used **functions** to group actions and **dictionaries** to group related data. But what if you want to group both **data + behavior** together in one structure? That's where **classes** come in.

### What is a Class?
A **class** is like a **blueprint** for creating **objects**. Each object (or **instance**) will have:
- Its own data (like a dictionary)
- Its own functions (called **methods**) Think of it like designing your own `type` of thing!

````python
#  Example: A Simple Pet
class Pet:
  def __init__(self, name, animalType):
    self.name = name
    self.animalType = animalType
    self.energy = 5

  def play(self):
    if self.energy > 0:
      print(self.name, "is playing!")
      self.energy -= 1
    else:
      print(self.name, "is too tired to play.")

  def feed(self):
    print(self.name, "is eating...")
    self.energy += 2

  def __repr__(self):
    return f"My {self.animalType}'s name is {self.name}"

  def __len__(self):
    return 1
````

````python
buddy = Pet("Buddy", "Golden Retriever")
print(buddy)
buddy.play()
buddy.feed()
print(buddy.name, f"is only {len(buddy)}", buddy.animalType)
print("The DataType of", buddy.name, "is", type(buddy))
````

#### Quick Exercise: Student
You already know how to:

- Write functions to organize code

- Use a dictionary to store related information about one student

- Use a list to store multiple students

The next step is to make this cleaner and easier to manage by turning it into a class. Instead of juggling functions, dicts, and lists separately, we can put everything that belongs to a student inside a single object.

````python
class Student:
    def __init__(self, name, age, grades):
        self.name = name
        self.age = age
        self.grades = grades

    def calculate_average(self):
        return sum(self.grades.values()) / len(self.grades)

    def assign_letter_grade(self):
        avg = self.calculate_average()
        if avg >= 90:
            return "A"
        elif avg >= 80:
            return "B"
        elif avg >= 70:
            return "C"
        elif avg >= 60:
            return "D"
        else:
            return "F"

    def print_report(self):
        print(f"Report for {self.name} (Age {self.age}):")
        for subject, grade in self.grades.items():
            print(f"  {subject}: {grade}")
        print(f"  Average: {self.calculate_average():.2f}")
        print(f"  Final Grade: {self.assign_letter_grade()}")
        print()


students = [
    Student("Mina", 16, {"Math": 95, "English": 88, "Science": 92}),
    Student("Omar", 17, {"Math": 75, "English": 82, "Science": 70}),
    Student("Sara", 16, {"Math": 60, "English": 65, "Science": 58}),
]

for student in students:
    student.print_report()
````
