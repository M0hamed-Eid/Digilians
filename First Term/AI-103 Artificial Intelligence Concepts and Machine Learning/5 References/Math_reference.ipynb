```python
import numpy as np
import matplotlib.pyplot as plt

```

```python
v_3d = np.array([25, 175, 70]) #3D Vector (age, height, weight)

# Create a 3D plot
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# Plot the vector from origin (0,0,0)
ax.quiver(
    0, 0, 0,                 # origin
    v_3d[0], v_3d[1], v_3d[2],  # vector components
    arrow_length_ratio=0.1
)

# Set axis labels
ax.set_xlabel("X")
ax.set_ylabel("Y")
ax.set_zlabel("Z")

# Set plot limits
ax.set_xlim([0, 25])
ax.set_ylim([0, 200])
ax.set_zlim([0, 100])

ax.set_title("3D Vector Representation")
plt.show()
```

```python
u = np.array([-1, 0, 1])
v = np.array([5, 2, -7])
print(f"The Dot Product of u & v = {np.dot(u,v)}") # -1*5 + 0*2 + 1*-7
print(f"The Cross Product of u & v = {np.cross(u,v)}")
```

```python
u_length = np.linalg.norm(u) # sqrt[(-1)^2 + (0)^2 + (1)^2]
v_length = np.linalg.norm(v) # sqrt[(5)^2 + (2)^2 + (-7)^2]
print(f"The Length of u = {u_length}")
print(f"The Length of v = {v_length}")
```

```python
u2 = np.array([1, 4])
v2 = np.array([2, -0.5])
print(f"The Dot Product of u2 & v2 = {np.dot(u2,v2)}") #orthogonal Vectors
# Plot vectors from origin
plt.quiver(0, 0, u2[0], u2[1], scale=1)
plt.quiver(0, 0, v2[0], v2[1], scale=1)
plt.grid()
plt.show()
```

```python
A = np.array([[1, 2],
              [3, 4]])

B = np.array([[5, 6],
              [7, 8]])
 
A + B   # [[1+5,  2+6],
        #  [7+3,  8+4]]
```

```python
# Matrix Vector Mult.
x = np.array([3, 4])
A @ x                # 1*3+4*2 , 3*3+4*4 ---> 11,25
```

```python
square = np.array([
    [0, 0],
    [1, 0],
    [1, 1],
    [0, 1],
    [0, 0]  ])

M = np.array([
    [1.2, 0.8],
    [-0.6, 0.4] ]) 
# Different magnitudes in each row -> non-uniform scaling
# Mixed positive and negative signs -> reflection
# Non-zero off-diagonal values -> shear

transformed_square = square @ M.T

plt.plot(square[:,0], square[:,1])
plt.plot(transformed_square[:,0], transformed_square[:,1])
plt.axis("equal")
plt.grid(True)
plt.show()
```
