# Python Implementation Of Q-Q Plot

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate example data
np.random.seed(0)
data = np.random.normal(size=1000)

# Create Q-Q plot
stats.probplot(data, plot=plt)
plt.title('Normal Q-Q plot')
plt.xlabel('Theoretical quantiles')
plt.ylabel('Ordered Values')
plt.grid(True)
plt.show()
```

```python
import numpy as np
import matplotlib.pyplot as plt
import scipy.stats as stats

# Generate a random sample from a normal distribution
normal_data = np.random.normal( size=1000)

# Generate a random sample from a right-skewed distribution (exponential distribution)
right_skewed_data = np.random.exponential(size=1000)

# Generate a random sample from a left-skewed distribution (negative exponential distribution)
left_skewed_data = -np.random.exponential( size=1000)

# Generate a random sample from an under-dispersed distribution (truncated normal distribution)
under_dispersed_data = np.random.normal(loc=0, scale=0.5, size=1000)
under_dispersed_data = under_dispersed_data[(under_dispersed_data > -1) & (under_dispersed_data < 1)]  # Truncate

# Generate a random sample from an over-dispersed distribution (mixture of normals)
over_dispersed_data = np.concatenate((np.random.normal(loc=-2, scale=1, size=500),
                                      np.random.normal(loc=2, scale=1, size=500)))

# Create Q-Q plots
plt.figure(figsize=(15, 10))

plt.subplot(2, 3, 1)
stats.probplot(normal_data, dist="norm", plot=plt)
plt.title('Q-Q Plot - Normal Distribution')

plt.subplot(2, 3, 2)
stats.probplot(right_skewed_data, dist="expon", plot=plt)
plt.title('Q-Q Plot - Right-skewed Distribution')

plt.subplot(2, 3, 3)
stats.probplot(left_skewed_data, dist="expon", plot=plt)
plt.title('Q-Q Plot - Left-skewed Distribution')

plt.subplot(2, 3, 4)
stats.probplot(under_dispersed_data, dist="norm", plot=plt)
plt.title('Q-Q Plot - Under-dispersed Distribution')

plt.subplot(2, 3, 5)
stats.probplot(over_dispersed_data, dist="norm", plot=plt)
plt.title('Q-Q Plot - Over-dispersed Distribution')

plt.tight_layout()
plt.show()
```

#  Histogram in Python

```python
import numpy as np

data = [5, 7, 10, 12, 15, 18, 20, 22, 25, 30, 35, 40]

import matplotlib.pyplot as plt

plt.hist(data, bins=5)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram with Equal Class Interval")
plt.show()
```

# Specify Exact Equal Intervals

```python
bins = [0, 10, 20, 30, 40, 50]

plt.hist(data, bins=bins)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Equal Class Interval Histogram")
plt.show()
```

# Histogram with Unequal Class Intervals

```python
bins = [0, 10, 20, 35, 50]

plt.hist(data, bins=bins)
plt.xlabel("Value")
plt.ylabel("Frequency")
plt.title("Histogram with Unequal Class Interval")
plt.show()
```

```python
plt.hist(data, bins=bins, density=True)
plt.xlabel("Value")
plt.ylabel("Frequency Density")
plt.title("Unequal Class Interval Histogram (Correct)")
plt.show()
```

# Histogram & Q–Q Plot on Dataset

```python
import pandas as pd

# Load Kaggle Titanic dataset (after downloading)
df = pd.read_csv("train.csv")

# Select a numeric column
age = df['Age'].dropna()
```

# Histogram (Distribution Check), Purpose
#See shape
#Detect skewness
#Identify outliers

```python
import matplotlib.pyplot as plt

plt.hist(age)
plt.xlabel("Age")
plt.ylabel("Frequency")
plt.title("Histogram of Age (Titanic Dataset)")
plt.show()
```

```python
s = pd.Series(age)

skewness = s.skew()
skewness
```

# Q–Q Plot (Normality Check)
#Purpose
#Check if data follows normal distribution
#Points on straight line → normal

```python
from scipy import stats

stats.probplot(age, dist="norm", plot=plt)
plt.title("Q–Q Plot of Age (Titanic Dataset)")
plt.show()
```

```python
#Points deviate at tails
#Data is not normally distributed
```

# Covariance & Correlation

```python
import pandas as pd

df = pd.read_csv("train.csv")

# Select relevant columns and drop missing values
data = df[['Age', 'Fare']].dropna()
cov_matrix = data.cov()
print(cov_matrix)
```

```python
#Positive covariance → Age and Fare increase together
#Value magnitude has no direct meaning
```

```python
corr_matrix = data.corr()
print(corr_matrix)
```

```python
#Weak positive relationship
```

```python
cov_age_fare = data['Age'].cov(data['Fare'])
corr_age_fare = data['Age'].corr(data['Fare'])

print("Covariance:", cov_age_fare)
print("Correlation:", corr_age_fare)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

sns.scatterplot(x='Age', y='Fare', data=data)
plt.title("Age vs Fare (Titanic Dataset)")
plt.show()
```
