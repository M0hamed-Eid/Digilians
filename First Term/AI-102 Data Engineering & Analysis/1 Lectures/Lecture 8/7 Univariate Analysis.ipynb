# 1️⃣ Mean for Ungrouped Data

```python
data = [10, 20, 30, 40, 50]
```

```python
mean_value = sum(data) / len(data)
print(mean_value)
```

#  Using NumPy

```python
import numpy as np

mean_value = np.mean(data)
print(mean_value)
```

# Using Pandas

```python
import pandas as pd

s = pd.Series(data)
print(s.mean())
```

# Mean for Grouped Data

# Grouped data means values with frequencies.

```python
data = {
    'class': ['A', 'A', 'B', 'B', 'B'],
    'score': [70, 80, 60, 75, 85]
}

df = pd.DataFrame(data)
```

```python
group_mean = df.groupby('class')['score'].mean()
print(group_mean)
```

```python
import numpy as np

values = [10, 20, 30]
weights = [2, 3, 5]

weighted_mean = np.average(values, weights=weights)
print(weighted_mean)
#         avg = sum(a * weights) / sum(weights)
```

# Geometric Mean (GM)

```python
from scipy.stats import gmean

data = [2, 4, 8, 16]

gm = gmean(data)
print(gm)
```

# Harmonic Mean (HM)

```python
from scipy.stats import hmean

data = [10, 20, 40]

hm = hmean(data)
print(hm)
```

#  Median (Ungrouped)

```python
import numpy as np

median_value = np.median(data)
print(median_value)
```

# Mode (Ungrouped)

```python
import pandas as pd
data = [10, 20, 40, 40, 20]
mode_value = pd.Series(data).mode()
#print(mode_value.tolist())
mode_value
```

# Grouped Data

```python
import numpy as np

values = [10, 20, 30]
freq = [2, 3, 5]

expanded = np.repeat(values, freq)
median_grouped = np.median(expanded)
print(median_grouped)
```

```python
import pandas as pd

df = pd.DataFrame({'x': values, 'f': freq})
df
```

```python

mode_grouped = df.loc[df['f'].idxmax(), 'x']
print(mode_grouped)
```

# Range 

```python
data = [10, 20, 30, 40, 50]
```

```python
data_range = max(data) - min(data)
print(data_range)
```

```python
import numpy as np

# Population variance
var_population = np.var(data)

# Sample variance
var_sample = np.var(data)

print(var_population, var_sample)
```

```python
import pandas as pd

s = pd.Series(data)

# Sample variance (default)
print(s.var())
```

```python
# Population STD
std_population = np.std(data)

# Sample STD
std_sample = np.std(data, ddof=1)

print(std_population, std_sample)
```

```python
print(s.std())
```

# How to Calculate Skewness and Kurtosis python

```python
data = [10, 12, 15, 18, 20, 100]
```

```python
import pandas as pd
data = [10, 12, 15, 18, 20, 100]

s = pd.Series(data)

skewness = s.skew()
kurtosis = s.kurt()   #  (excess kurtosis)

print(skewness, kurtosis)
```
