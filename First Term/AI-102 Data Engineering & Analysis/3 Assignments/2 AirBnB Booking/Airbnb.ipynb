```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
%matplotlib inline
import seaborn as sns
import warnings
warnings.filterwarnings('ignore')
```

```python
Airbnb_df = pd.read_csv('Airbnb_Open_Data.csv')
Airbnb_df
```

```python
rename_col = {'id':'listing_id','name':'listing_name','number_of_reviews':'total_reviews','calculated_host_listings_count':'host_listings_count'}
```

```python
# use a pandas function to rename the current function
Airbnb_df = Airbnb_df.rename(columns = rename_col)
Airbnb_df.head(2)
```

```python
#checking shape of Airbnb dataset
Airbnb_df.shape
```

```python
#basic information about the dataset
Airbnb_df.info()
```

```python
# check duplicate rows in dataset
Airbnb_df = Airbnb_df.drop_duplicates()
Airbnb_df.count()
```

```python
# checking null values of each columns
Airbnb_df.isnull().sum()
```

```python
# Step 1: Clean the 'price' and 'service fee' columns (remove '$' and convert to numeric)
Airbnb_df['price'] = Airbnb_df['price'].replace({'\$': '', ',': ''}, regex=True).astype(float)
Airbnb_df['service fee'] = Airbnb_df['service fee'].replace({'\$': '', ',': ''}, regex=True).astype(float)
```

```python
# Step 2: Handle missing values in numerical columns (filling with median)
numerical_columns = ['lat', 'long', 'price', 'service fee', 'minimum nights',
                     'number of reviews', 'reviews per month', 'review rate number',
                     'calculated host listings count', 'availability 365']

for col in numerical_columns:
    Airbnb_df[col] = Airbnb_df[col].fillna(Airbnb_df[col].median())
```

```python
# Step 3: Handling missing values in categorical columns (filling with mode)
categorical_columns = ['NAME', 'host_identity_verified', 'host name', 
                       'neighbourhood group', 'neighbourhood', 'country', 
                       'country code', 'instant_bookable', 'cancellation_policy', 
                       'room type', 'Construction year', 'house_rules', 'license']

for col in categorical_columns:
    Airbnb_df[col] = Airbnb_df[col].fillna(Airbnb_df[col].mode()[0])
```

```python
# Step 4: Drop columns with excessive missing data
Airbnb_df.drop(columns=['house_rules', 'license'], inplace=True)
```

```python
# Step 5: One-hot encoding categorical variables
Airbnb_df = pd.get_dummies(Airbnb_df, columns=['neighbourhood', 'country', 'room type'], drop_first=True)
```

```python
# Step 6: Check if there are still any null values left
Airbnb_df.isnull().sum()
```
