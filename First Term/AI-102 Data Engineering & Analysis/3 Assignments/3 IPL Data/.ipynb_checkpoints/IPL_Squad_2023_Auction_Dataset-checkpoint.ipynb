```python
import numpy as np
import pandas as pd
```

```python
data = pd.read_csv('IPL_Squad_2023_Auction_Dataset.csv')
data.sample(5)
```

```python
data.info()
```

```python
data.describe()
```

```python
# Step 1: Drop unnecessary column
data.drop(columns=['Unnamed: 0'], inplace=True)
```

```python
# Step 2: Clean 'Base Price'

# Replace 'Retained' with 0
data['Base Price'] = data['Base Price'].replace('Retained', 0)

# Convert to numeric
data['Base Price'] = pd.to_numeric(data['Base Price'])
```

```python
# Step 3: Fill missing values in cost columns with 0
data['COST IN ₹ (CR.)'] = data['COST IN ₹ (CR.)'].fillna(0)
data['Cost IN $ (000)'] = data['Cost IN $ (000)'].fillna(0)
```

```python
# Step 4: Fill missing values in '2022 Squad'
data['2022 Squad'] = data['2022 Squad'].fillna('Not in 2022 Squad')
```

```python
# Step 5: Final null check
data.isnull().sum()
```

```python
# Step 6: One-hot encoding categorical columns
data = pd.get_dummies(data, columns=['TYPE', 'Team'], drop_first=True)
```

```python
# Step 7: Final dataset info
data.info()
```

```python
data.to_csv("cleaned_IPL_2023_dataset.csv", index=False)
```
