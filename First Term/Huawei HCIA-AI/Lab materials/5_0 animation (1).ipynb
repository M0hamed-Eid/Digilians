```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, FFMpegWriter
from IPython.display import HTML

# Sample data
np.random.seed(0)
X = np.linspace(0, 10, 500)
y = 2 * X + 1 + np.random.normal(scale=3, size=X.shape)

# Gradient descent parameters
theta0, theta1 = -1, 10
lr = 0.01
epochs = 20

# Store parameter updates
theta0_history = [theta0]
theta1_history = [theta1]

for _ in range(epochs):
    y_pred = theta0 + theta1 * X
    error = y_pred - y
    theta0 -= lr * error.mean()
    theta1 -= lr * (error * X).mean()
    theta0_history.append(theta0)
    theta1_history.append(theta1)

# Create figure
fig, ax = plt.subplots()
ax.scatter(X, y, color='blue', label='Data points')
line, = ax.plot([], [], color='red', lw=2, label='Regression line')
residuals = [ax.plot([], [], color='green', lw=1)[0] for _ in range(len(X))]

ax.set_xlim(X.min()-1, X.max()+1)
ax.set_ylim(y.min()-5, y.max()+5)
ax.set_xlabel('X')
ax.set_ylabel('y')
ax.legend()

# Animation function
def animate(i):
    y_line = theta0_history[i] + theta1_history[i] * X
    line.set_data(X, y_line)
    for j, res_line in enumerate(residuals):
        res_line.set_data([X[j], X[j]], [y[j], y_line[j]])
    ax.set_title(f"Linear Regression Fit - Step {i}")
    return [line] + residuals

anim = FuncAnimation(fig, animate, frames=len(theta0_history), interval=2000)

# Display as video in notebook
HTML(anim.to_html5_video())
```
