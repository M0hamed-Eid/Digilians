```python
import numpy as np

np.random.seed(42)  # to make this code example reproducible
m = 100  # number of instances
X = 2 * np.random.rand(m, 1)  # column vector
y = 4 + 3 * X + np.random.randn(m, 1)  # column vector
```

```python
X
```

```python
from sklearn.preprocessing import add_dummy_feature

X_b = add_dummy_feature(X)  # add x0 = 1 to each instance
```

```python
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
theta_best
```

```python
X_new = np.array([[0]])
X_new_b = add_dummy_feature(X_new)  # add x0 = 1 to each instance
y_predict = X_new_b @ theta_best
y_predict
```

```python
eta = 0.1  # learning rate
n_epochs = 1000
m = len(X_b)  # number of instances

np.random.seed(42)
theta = np.random.randn(2, 1)  # randomly initialized model parameters

for epoch in range(n_epochs):
    gradients = 2 / m * X_b.T @ (X_b @ theta - y)
    theta = theta - eta * gradients
```

```python
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.animation import FuncAnimation, PillowWriter

# Sample data
np.random.seed(0)
X = np.linspace(0, 10, 100)
y = 2 * X + 1 + np.random.normal(scale=3, size=X.shape)

# Gradient descent parameters
theta0, theta1 = 0, 0
lr = 0.01
epochs = 50

theta0_history = [theta0]
theta1_history = [theta1]

for _ in range(epochs):
    y_pred = theta0 + theta1 * X
    error = y_pred - y
    theta0 -= lr * error.mean()
    theta1 -= lr * (error * X).mean()
    theta0_history.append(theta0)
    theta1_history.append(theta1)

# Figure
fig, ax = plt.subplots()
ax.scatter(X, y, color='blue', label='Data points')
line, = ax.plot([], [], color='red', lw=2, label='Regression line')
ax.set_xlim(X.min()-1, X.max()+1)
ax.set_ylim(y.min()-5, y.max()+5)
ax.set_xlabel('X')
ax.set_ylabel('y')
ax.legend()

def animate(i):
    line.set_data(X, theta0_history[i] + theta1_history[i] * X)
    ax.set_title(f"Linear Regression Fit - Step {i}")
    return line,

anim = FuncAnimation(fig, animate, frames=len(theta0_history), interval=100)

# Option 1: Save as GIF (or mp4)
anim.save("linear_regression.gif", writer=PillowWriter(fps=10))
print("Animation saved as linear_regression.gif")

# Option 2: If in Jupyter Notebook, display inline
%matplotlib notebook
from IPython.display import HTML
HTML(anim.to_jshtml())
```
