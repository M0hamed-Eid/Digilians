```python
import numpy as np
```

```python
np.random.seed(42)
```

```python
m=100
```

```python
X = 2*np.random.rand(m,3)
X.shape
```

```python
y = 4 + (3*X[:,0])+(2*X[:,1])+(0.5*X[:,2])+  np.random.randn(m,1).flatten()
```

```python
X
```

```python
import matplotlib.pyplot as plt
```

```python
plt.scatter(X,y)
```

```python
from sklearn.preprocessing import add_dummy_feature
```

```python
X_b = add_dummy_feature(X)
```

```python
X_b.shape
```

```python
X_b[0]
```

```python
X_b
```

```python
X_b.shape#feature matrix
```

```python
y.shape#target array, groudtruth, labels
```

```python
y.shape
```

```python
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```
