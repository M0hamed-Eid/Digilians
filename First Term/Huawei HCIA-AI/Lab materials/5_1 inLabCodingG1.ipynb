**Normal Equation(Closed Form)**

```python
import numpy as np

np.random.seed(42)
m = 100
X = 2*np.random.rand(m,1)
X.shape
X[0]
y = 4 + 3*X + np.random.randn(m,1)
```

```python
y = 4 + 3*X + np.random.randn(m,1)
```

```python
X.shape
```

```python
y.shape
```

```python
import matplotlib.pyplot as plt
plt.scatter(X,y)
```

```python
from sklearn.preprocessing import add_dummy_feature
X_b = add_dummy_feature(X)
X_b.shape
X_b[0]
```

```python
xb = np.concatenate([np.ones((m,1)),X],axis=1)
xb.shape
xb[0]
```

```python
theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
```

```python
y = 4 + 3*X + np.random.randn(m,1)
```

```python
theta_best
```

```python
import numpy as np

np.random.seed(42)
m = 100
X = 2*np.random.rand(m,3)
X.shape
X[0]
y = 4+3*X[:,0]+2*X[:,1]+0.5*X[:,2] + np.random.randn(m,1).ravel()


X_b = add_dummy_feature(X)

theta_best = np.linalg.inv(X_b.T @ X_b) @ X_b.T @ y
theta_best
```

```python
X.shape
```

```python
y = 4+3*X[:,0]+2*X[:,1]+0.5*X[:,2]  + np.random.randn(m,1).ravel()
y.shape
```
