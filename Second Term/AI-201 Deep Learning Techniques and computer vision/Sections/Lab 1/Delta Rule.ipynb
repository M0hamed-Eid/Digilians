```python
import numpy as np

# =====================================================
# 1) AND DATASET
# =====================================================
# Input features (x1, x2)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Target output for AND gate
# Output = 1 only if both inputs are 1
y = np.array([0, 0, 0, 1])


# =====================================================
# 2) INITIALIZE WEIGHTS AND PARAMETERS
# =====================================================
np.random.seed(42)

# Initialize weights (one weight per input)
W = np.random.randn(2)

# Initialize bias
b = np.random.randn()

# Learning rate
learning_rate = 0.1

# Number of training iterations
epochs = 20


# =====================================================
# 3) ACTIVATION FUNCTION (STEP FUNCTION)
# =====================================================
# Converts continuous value into binary output (0 or 1)
def step_function(z):
    return 1 if z >= 0 else 0


# =====================================================
# 4) TRAINING USING DELTA RULE
# =====================================================
for epoch in range(epochs):

    print(f"\nEpoch {epoch+1}")

    for i in range(len(X)):

        # -------------------------------
        # FORWARD PASS
        # -------------------------------
        # Linear combination: z = w.x + b
        z = np.dot(X[i], W) + b

        # Apply step function to get prediction
        y_hat = step_function(z)

        # -------------------------------
        # ERROR CALCULATION
        # -------------------------------
        # Error = true output - predicted output
        error = y[i] - y_hat

        # -------------------------------
        # DELTA RULE UPDATE
        # -------------------------------
        # Update weights
        W = W + learning_rate * error * X[i]

        # Update bias
        b = b + learning_rate * error

        print(f"Input: {X[i]}, Predicted: {y_hat}, Error: {error}")

    print(f"Weights: {W}, Bias: {b}")


# =====================================================
# 5) FINAL TEST
# =====================================================
print("\nFinal Predictions:")

for i in range(len(X)):
    z = np.dot(X[i], W) + b
    y_hat = step_function(z)
    print(f"Input: {X[i]} -> Output: {y_hat}")
```
