```python
import numpy as np

# =====================================================
# 1) AND DATASET
# =====================================================
# Input data (all possible combinations)
X = np.array([
    [0, 0],
    [0, 1],
    [1, 0],
    [1, 1]
])

# Target outputs for AND gate
y = np.array([0, 0, 0, 1])
```

```python
# =====================================================
# 2) INITIALIZE PARAMETERS
# =====================================================
np.random.seed(42)

# Initialize weights
W = np.random.randn(2)

# Initialize bias
b = np.random.randn()

# Number of training iterations
epochs = 20


# =====================================================
# 3) STEP ACTIVATION FUNCTION
# =====================================================
def step_function(z):
    return 1 if z >= 0 else 0

```

```python

# =====================================================
# 4) PERCEPTRON TRAINING (NO LEARNING RATE)
# =====================================================
for epoch in range(epochs):

    print(f"\nEpoch {epoch+1}")

    for i in range(len(X)):

        # -------------------------------
        # FORWARD PASS
        # -------------------------------
        z = np.dot(X[i], W) + b
        y_hat = step_function(z)

        # -------------------------------
        # ERROR CALCULATION
        # -------------------------------
        error = y[i] - y_hat

        # -------------------------------
        # UPDATE RULE (learning rate = 1)
        # -------------------------------
        # Weight update
        W += error * X[i]

        # Bias update
        b += error

        print(f"Input: {X[i]}, Predicted: {y_hat}, Error: {error}")

    print(f"Weights: {W}, Bias: {b}")


# =====================================================
# 5) FINAL PREDICTION
# =====================================================
print("\nFinal Predictions:")

for i in range(len(X)):
    z = np.dot(X[i], W) + b
    y_hat = step_function(z)
    print(f"Input: {X[i]} → Output: {y_hat}")
```
