# Perceptron

```python
import numpy as np

X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([0,1,1,1])

# weights initialization
w1 = 0.1
w2 = 0.2
w_bias = 1

threshold = 0.1

epochs = 10

for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    
    for i in range(len(X)):
        x1 = X[i][0]
        x2 = X[i][1]
        bias = 1

        # net calculation
        net = x1*w1 + x2*w2 + bias*w_bias

        # activation
        y_pred = 1 if net >= threshold else 0

        t = y[i]

        # update rules exactly as you wrote
        if y_pred != t:
            
            # case 1: y = 0, want y = 1 (increase weights)
            if y_pred == 0 and t == 1:
                w1 = w1 + x1
                w2 = w2 + x2
                w_bias = w_bias + bias

            # case 2: y = 1, want y = 0 (decrease weights)
            elif y_pred == 1 and t == 0:
                w1 = w1 - x1
                w2 = w2 - x2   
                w_bias = w_bias - bias

        print(f"x1={x1}, x2={x2}, target={t}, pred={y_pred}, w1={w1}, w2={w2}, wb={w_bias}")
```

# SLP with Single output

```python
import numpy as np

X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([0,1,1,1])

# initialize weights
w1 = 0.1
w2 = 0.2
wb = -0.2

learning_rate = 0.1
threshold = 0.1
epochs = 10

for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")
    
    for i in range(len(X)):
        x1 = X[i][0]
        x2 = X[i][1]
        bias = 1

        # net input
        net = x1*w1 + x2*w2 + bias*wb

        # activation
        y_pred = 1 if net >= threshold else 0

        t = y[i]

        # error
        error = t - y_pred

        # SLP update rule 
        w1 = w1 + learning_rate * x1 * error
        w2 = w2 + learning_rate * x2 * error
        wb = wb + learning_rate * bias * error

        print(f"x1={x1}, x2={x2}, t={t}, y={y_pred}, w1={w1}, w2={w2}, wb={wb}")
```

# SLP with multiple outputs

```python
import numpy as np

X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([[0,0],
              [1,0],
              [1,0],
              [1,1]])

learning_rate = 0.1

# weights (as given)
w11, w21, wb1 = 0.2, 0.3, 0.0
w12, w22, wb2 = 0.1, 0.2, 0.7

threshold1 = 0.1
threshold2 = 1

epochs = 10

for epoch in range(epochs):
    print(f"\nEpoch {epoch+1}")

    for i in range(len(X)):
        x1, x2 = X[i]
        bias = 1

        # ---------- OUTPUT 1 ----------
        net1 = x1*w11 + x2*w21 + bias*wb1
        y1 = 1 if net1 >= threshold1 else 0
        t1 = y[i][0]
        e1 = t1 - y1

        w11 = w11 + learning_rate * x1 * e1
        w21 = w21 + learning_rate * x2 * e1
        wb1 = wb1 + learning_rate * bias * e1

        # ---------- OUTPUT 2 ----------
        net2 = x1*w12 + x2*w22 + bias*wb2
        y2 = 1 if net2 >= threshold2 else 0
        t2 = y[i][1]
        e2 = t2 - y2

        w12 = w12 + learning_rate * x1 * e2
        w22 = w22 + learning_rate * x2 * e2
        wb2 = wb2 + learning_rate * bias * e2

        print(f"x=({x1},{x2}) | "
            f"actual=({t1},{t2}) | "
            f"pred=({y1},{y2}) | "
            f"w1=({w11:.2f},{w21:.2f},{wb1:.2f}) | "
            f"w2=({w12:.2f},{w22:.2f},{wb2:.2f})")
```

# Perceptron Keras

```python
import numpy as np
import tensorflow as tf

X = np.array([[0,0],
              [0,1],
              [1,0],
              [1,1]])

y = np.array([0,1,1,1])

model = tf.keras.Sequential([
    tf.keras.layers.Dense(1, input_shape=(2,), activation='sigmoid')
])

model.compile(loss='mse')

model.fit(X, y, epochs=10, verbose=1)
```

```python
print(model.predict(X))
```

```python
print(model.predict(X).round())
```

# Multi-Output SLP Keras
